## Explore Datasets
Use the `descriptions`, `films`, `people`, `reviews`, and `roles` tables to explore the data and practice your skills!
- Which titles in the `reviews` table have an IMDB score higher than 8.5?
- Select all titles from Germany released after 2010 from the `films` table.

### Titles in the reviews table with higher IMDB SCORE than 8.5

```python
SELECT film_id, imdb_score
FROM cinema.reviews
WHERE imdb_score > 8.5;
```

### Titles from Germany released after 2010

```python
SELECT title
FROM cinema.films
WHERE language = 'German'
	AND release_year in (1999, 2000, 2001);
```

### Calculate a number of all movies by country using the `films` table.

```python
SELECT country, COUNT(*) AS number_of_movies
FROM cinema.films
WHERE country is not null
GROUP BY country;
```

```python
SELECT COUNT(DISTINCT(country)) as count_distinct_countries
FROM cinema.films
WHERE country is not null
```

```python
-- German films titles with release year that we do not know the budget.
SELECT title AS no_budget_info, release_year
FROM cinema.films
WHERE language = 'German'
	AND budget IS NULL;
```

```python
SELECT title, budget
FROM cinema.films
WHERE budget is not null
ORDER BY cinema.films.budget DESC;
```

```python
SELECT ROUND(AVG(budget), -3) AS avg_budget
FROM cinema.films
WHERE release_year >= 2010;
```

```python
SELECT MAX(budget) AS max_budget, 
       MAX(duration) AS max_duration
FROM cinema.films 
WHERE cinema.films.release_year < 2000;
```

```python
SELECT title, ROUND(duration / 60.0, 2) AS duration_hours
FROM cinema.films
WHERE cinema.films.release_year < 2000
ORDER BY duration_hours DESC;
```

```python
-- Films are based on the 10 decades.
SELECT (MAX(release_year) - MIN(release_year)) / 10.0 AS number_of_decades
FROM cinema.films;
```

Top five American film titles released between 2004-2006 with the longest duration of projection (by minutes).

```python
SELECT title, release_year, duration
FROM cinema.films
WHERE cinema.films.country = 'USA'
	AND release_year BETWEEN 2004 AND 2006
	AND duration IS NOT NULL
ORDER BY duration DESC
LIMIT 6;
```

Number of films release each year.

Maximum budget by country and by release year.

```python
SELECT release_year, country, MAX(budget) AS max_budget
FROM cinema.films
WHERE cinema.films.budget IS NOT NULL
GROUP BY cinema.films.release_year, cinema.films.country
ORDER BY cinema.films.release_year, cinema.films.country;
```

Which release year had the most diversity by language so far ?
It seems that it is 2006 - by so far 17 languages.

```python
SELECT release_year, COUNT(DISTINCT language) AS diversity
FROM cinema.films
WHERE cinema.films.language IS NOT NULL
GROUP BY cinema.films.release_year
ORDER BY diversity DESC;
```

In what years was the average film duration over two hours?

```python
SELECT release_year
FROM cinema.films
GROUP BY cinema.films.release_year
HAVING AVG(cinema.films.duration) > 120;
```

Country with number of distinct certification of films greater than 10.

```python
SELECT country, COUNT(DISTINCT certification) AS certification_count
FROM cinema.films
GROUP BY cinema.films.country
HAVING COUNT(DISTINCT cinema.films.certification) > 10;
```

Countries with an average budget of more than one billion in descending order.

```python
SELECT country, AVG(budget) AS average_budget
FROM cinema.films
GROUP BY cinema.films.country
HAVING AVG(budget) > 1000000000
ORDER BY average_budget DESC;
```

Release year of film after 1990 with average budget is more than 60 milions, and it is ordered by average gross earning.
The average budget and gross earnings for films each year after 1990 if the average budget is greater than 60 million.

```python
SELECT release_year, AVG(budget) AS avg_budget, AVG(gross) AS avg_gross
FROM cinema.films
WHERE cinema.films.release_year > 1990
GROUP BY cinema.films.release_year
HAVING AVG(budget) > 60000000
ORDER BY avg_gross DESC
LIMIT 1;
```
